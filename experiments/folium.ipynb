```python
from folium import Map
from folium.plugins import HeatMap
import json
import pandas as pd
```

```python
# Get data
f = open("ovation_aurora_latest.json")
data = json.loads(f.read())
# Change order to Lat, Lon, Weight
coords = [[row[1], row[0], row[2]] for row in data['coordinates']]
#df = pd.DataFrame(data=data['coordinates'], columns=["Longitude", "Latitude", "Aurora"])
#df = df[["Latitude", "Longitude", "Aurora"]]
#df.values.tolist() == coords
```

```python
def generateBaseMap(default_location=[64.9648751621697,17.6754094331351], default_zoom_start=5):
    base_map = Map(location=default_location, control_scale=True, zoom_start=default_zoom_start)
    return base_map
```

```python
base_map = generateBaseMap()
#HeatMap(data=coords, radius=10, min_opacity=10, max_zoom=20).add_to(base_map)
HeatMap(data=coords, min_opacity=0, blur=50, radius=5).add_to(base_map)
base_map
```
